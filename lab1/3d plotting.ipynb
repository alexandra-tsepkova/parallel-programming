```python
l = list(map(str.split, open('data_serial.txt'))) # insert filename here
for j in range(len(l)):
    l[j] = [float(i) for i in l[j]]
l = np.array(l)
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

t = np.linspace(0, 1, len(l))
x = np.linspace(0, 5, len(l[0]))
x, t = np.meshgrid(x, t)
zdata = open("data.txt", "r")
surf = ax.plot_surface(x, t, l, cmap=cm.coolwarm,
                       linewidth=0)
ax.set_xlabel('x', fontsize=15)
ax.set_ylabel('t', fontsize=15)
ax.set_zlabel('u', fontsize=15)


plt.show()
```
